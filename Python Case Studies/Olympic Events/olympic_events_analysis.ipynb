```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("olympic_events.csv", encoding="ISO-8859-1")
data
```

```python
data["id"].unique().size
```

```python
data["sex"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(data["sex"].value_counts(), autopct="%1.2f%%", labels=data["sex"].value_counts().index)
plt.title("GENDER COUNT PIE CHART")
plt.show()
```

```python
mix_sex = data[data["sex"]=="mixed"]
mix_sex
```

```python
print("==COUNTS OF SPORTS HAVING MIX GENDER==")
print(mix_sex["sport"].value_counts())
sns.countplot(mix_sex["sport"])
plt.title("BAR CHART OF SPORTS HAVING MIX GENDER")
plt.show()
```
